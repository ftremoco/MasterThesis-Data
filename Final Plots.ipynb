```python
# Import required packages
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from pylab import cm
```

# 2D vs 3D plots

```python
acc5_3d = pd.read_csv('3d/acc5.txt')
acc5_2d = pd.read_csv('2d/acc5.txt')
erro = abs(acc5_3d['Pabs'] - acc5_2d['Pabs'])

# Edit the font, font size, and axes width
mpl.rcParams['font.family'] = 'Avenir'
plt.rcParams['font.size'] = 18
plt.rcParams['axes.linewidth'] = 2
mpl.rcParams['pdf.fonttype'] = 42

# Generate 2 colors from the 'tab10' colormap
colors = cm.get_cmap('tab10', 5)

# Create figure object and store it in a variable called 'fig'
fig = plt.figure(figsize=(15,5))


ax = fig.add_axes([0, 0, 0.5, 1])

# Add the x and y-axis labels
ax.set_xlabel('Wavelength (nm)', labelpad=10)
ax.set_ylabel('Fration of absorbed power ', labelpad=10)

# Set the axis limits
ax.set_xlim(340, 700)
ax.set_ylim(0, 1)

# Hide the top and right spines of the axis
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)


ax.plot(acc5_3d['lambda'],acc5_3d['Pabs'],'--', linewidth = 2, color='red', label='3D')
ax.plot(acc5_2d['lambda'],acc5_2d['Pabs'],'--',linewidth = 2, color='blue', label='2D')
ax.legend(bbox_to_anchor=(1, 1), loc=1, frameon=True, fontsize=16)



ax2 = fig.add_axes([0.6, 0, 0.4, 1])

# Add the x and y-axis labels
ax2.set_xlabel('Wavelength (nm)', labelpad=10)
ax2.set_ylabel('Fraction of absorbed power', labelpad=10)

#Hide the top and right spines of the axis
ax2.spines['right'].set_visible(False)
ax2.spines['top'].set_visible(False)

ax2.plot(acc5_3d['lambda'],erro,'.', linewidth = 2, color='green', label='ABS(3d-2d)')

ax2.legend(bbox_to_anchor=(1, 1), loc=1, frameon=True, fontsize=16)

#Save figure
plt.savefig('2dvs3d-plot.pdf', transparent=False, bbox_inches='tight')
```

# Wide-field: PNP-TR plots

```python
pnp1 = pd.read_csv('pnp-tr2d/acc1.txt')
pnp2 = pd.read_csv('pnp-tr2d/acc2.txt')
pnp3 = pd.read_csv('pnp-tr2d/acc3.txt')
pnp4 = pd.read_csv('pnp-tr2d/acc4.txt')
pnp5 = pd.read_csv('pnp-tr2d/acc5.txt')
pnp6 = pd.read_csv('pnp-tr2d/acc6.txt')
pnp7 = pd.read_csv('pnp-tr2d/acc7.txt')
pnp8 = pd.read_csv('pnp-tr2d/acc8.txt')

sigma1 = 1
sigma2 = abs(sum(pnp2['Pabs']-pnp1['Pabs'])/sum(pnp2['Pabs']))
sigma3 = abs(sum(pnp3['Pabs']-pnp2['Pabs'])/sum(pnp3['Pabs']))
sigma4 = abs(sum(pnp4['Pabs']-pnp3['Pabs'])/sum(pnp4['Pabs']))
sigma5 = abs(sum(pnp5['Pabs']-pnp4['Pabs'])/sum(pnp5['Pabs']))
sigma6 = abs(sum(pnp6['Pabs']-pnp5['Pabs'])/sum(pnp6['Pabs']))
sigma7 = abs(sum(pnp7['Pabs']-pnp6['Pabs'])/sum(pnp7['Pabs']))
sigma8 = abs(sum(pnp8['Pabs']-pnp7['Pabs'])/sum(pnp8['Pabs']))
sigma = np.array([sigma1,sigma2,sigma3,sigma4,sigma5,sigma6,sigma7,sigma8])


# Edit the font, font size, and axes width
mpl.rcParams['font.family'] = 'Avenir'
plt.rcParams['font.size'] = 18
plt.rcParams['axes.linewidth'] = 2
mpl.rcParams['pdf.fonttype'] = 42

# Generate 2 colors from the 'tab10' colormap
colors = cm.get_cmap('tab10', 8)

# Create figure object and store it in a variable called 'fig'
fig = plt.figure(figsize=(15,5))



ax = fig.add_axes([0, 0, 0.5, 1])

# Add the x and y-axis labels
ax.set_xlabel('Wavelength (nm)', labelpad=10)
ax.set_ylabel('Fraction of absorbed power ', labelpad=10)

# Set the axis limits
ax.set_xlim(340, 700)
ax.set_ylim(0, 1)

# Hide the top and right spines of the axis
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)


ax.plot(pnp1['lambda'],pnp1['Pabs'], linewidth = 2, color=colors(0), label='acc1')
ax.plot(pnp2['lambda'],pnp2['Pabs'], linewidth = 2, color=colors(1), label='acc2')
ax.plot(pnp3['lambda'],pnp3['Pabs'], linewidth = 2, color=colors(2), label='acc3')
ax.plot(pnp4['lambda'],pnp4['Pabs'], linewidth = 2, color=colors(3), label='acc4')
ax.plot(pnp5['lambda'],pnp5['Pabs'], linewidth = 2, color=colors(4), label='acc5')
ax.plot(pnp6['lambda'],pnp6['Pabs'], linewidth = 2, color=colors(5), label='acc6')
ax.plot(pnp7['lambda'],pnp7['Pabs'], linewidth = 2, color=colors(6), label='acc7')
ax.plot(pnp8['lambda'],pnp8['Pabs'], linewidth = 2, color=colors(7), label='acc8')
ax.legend(bbox_to_anchor=(1, 1), loc=1, frameon=True, fontsize=16)



ax2 = fig.add_axes([0.6, 0, 0.4, 1])

# Add the x and y-axis labels
ax2.set_xlabel('Mesh accuracy', labelpad=10)
ax2.set_ylabel('$$', labelpad=10)

#Hide the top and right spines of the axis
ax2.spines['right'].set_visible(False)
ax2.spines['top'].set_visible(False)


ax2.plot(np.arange(8)+1,sigma, 'v' , markersize = 7, color=colors(0), label='$$ (mesh accuracy)')
ax2.legend(bbox_to_anchor=(1, 1), loc=1, frameon=True, fontsize=16)

#Save figure
plt.savefig('pnptr-conv.pdf', transparent=False, bbox_inches='tight')

print(sigma)
```
